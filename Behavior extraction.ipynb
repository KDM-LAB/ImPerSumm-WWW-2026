```python
# prompt: so i want you to convert the Docs and Action columns to list and find the length of Docs and Action for each UserID

import pandas as pd
import ast

df = pd.read_csv('/content/synthetic-original_D2_25_200.csv')

def get_list_length(value):
  try:
    value_list = ast.literal_eval(value)
    if not isinstance(value_list, list):
      value_list = [value_list]
    return len(value_list)
  except (ValueError, SyntaxError):
    return 0  # Or handle the error as needed

df['Docs_Length'] = df['Docs'].apply(get_list_length)
df['Action_Length'] = df['Action'].apply(get_list_length)

print(df[['UserID', 'Docs_Length', 'Action_Length']])
```

```python
import pandas as pd
import ast
from tqdm import tqdm

# Read the CSV file containing user behavior data
df = pd.read_csv(r"C:\Users\admin\Desktop\ImPerSum\Datasets\train.csv")

# Initialize a list to store the edges and an edge counter for unique EdgeIDs
edges = []
edge_counter = 1

# Iterate over each user row in the DataFrame
for _, row in tqdm(df.iterrows(),total=len(df),desc='behavior extraction is being done.....'):
    user = row['UserID']

    # Convert the string representation of lists to actual lists using ast.literal_eval
    docs = ast.literal_eval(row['Docs'])
    actions = ast.literal_eval(row['Action'])

    # Check that there is at least one interaction
    if len(docs) > 0 and len(actions) > 0:
        # Create the first edge: head is the user id, relation is first action, tail is first doc id
        edge_id = f"B{edge_counter}"
        edges.append({
            "EdgeID": edge_id,
            "Head": user,
            "Relation": actions[0],
            "Tail": docs[0],
            "User": user
        })
        edge_counter += 1

        # For subsequent interactions, the head is the previous doc (tail)
        for i in range(1, len(docs)):
            edge_id = f"B{edge_counter}"
            edges.append({
                "EdgeID": edge_id,
                "Head": docs[i - 1],
                "Relation": actions[i],
                "Tail": docs[i],
                "User": user
            })
            edge_counter += 1

# Create a lookup table DataFrame from the edges list
lookup_df = pd.DataFrame(edges)

lookup_df
```

```python
results=[]
for user, user_df in tqdm(lookup_df.groupby('User'), desc='Processing Users'):
    user_df = user_df.reset_index(drop=True)
    
    # Track EdgeIDs for behaviors
    edge_ids = user_df['EdgeID'].tolist()
    relations = user_df['Relation'].tolist()
    
    bhist = []
    instance_counter = 1
    
    for idx, relation in enumerate(relations):
        if relation == 'summ_gen':
            bpos = edge_ids[idx]
            bhist_entry = edge_ids[:idx]  # all EdgeIDs before current one
            results.append({
                'UserID': f"{user}_{instance_counter}",
                'Bhist': bhist_entry,
                'Bpos': bpos
            })
            instance_counter += 1
```

```python
train_df = pd.DataFrame(results)
train_df
```

```python
train_df['Bpos'] = train_df['Bpos'].astype(str)
lookup_df['EdgeID'] = lookup_df['EdgeID'].astype(str)

# Filter lookup_df where EdgeID is in train_df['Bpos']
matching_rows = lookup_df[lookup_df['EdgeID'].isin(train_df['Bpos'])]

# Display the result
print(f"Total matching rows: {len(matching_rows)}")
matching_rows.head(20)
```
